```python
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
image = cv2.imread("Sample image.jpg")
```

```python
image_rgb = np.copy(image)
image_rgb = cv2.cvtColor(image_rgb,cv2.COLOR_BGR2RGB)
plt.imshow(image_rgb)
```

```python
image_gray = np.copy(image)
image_gray = cv2.cvtColor(image_gray,cv2.COLOR_BGR2GRAY) #image convert to gray
plt.imshow(image_gray,cmap='gray')
```

```python
image_gray.shape  #image shape in pixels
```

```python
cv2.minMaxLoc(image_gray) # this will return minimum value ,maximum value and their index
```

```python
max_loc_x = cv2.minMaxLoc(image_gray)[3][0] # x coordinate of pixel
max_loc_y = cv2.minMaxLoc(image_gray)[3][1] # ycoordinate of pixel
```

```python
s=70 
#starting and ending coordinates of rectangle
coord1= (max_loc_x-s,max_loc_y-s) 
coord2= (max_loc_x+s,max_loc_y+s)
#colour of rectangle
color=(255,255,255)
t=15 #thickness of rectangle

image_box = np.copy(image_rgb)
image_box = cv2.rectangle(image_box,coord1,coord2,color,t)
plt.imshow(image_box)
```
